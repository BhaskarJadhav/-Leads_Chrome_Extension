```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import time
```

```python
#define  2*2 enviorment

environment ={
     "Room1": "Clean",
     "Room2": "Clean",
     "Room3": "Dirty",
     "Room4": "Clean"
}
```

```python
#mqpping for the grid

room_positions ={
     "Room1": (0,1),   #top left
     "Room2": (1,1),   # top right
     "Room3": (0,0),   # bottom-left
     "Room4": (1,0)    # Bottom_right
}
```

```python
rooms = list(environment.keys())
agent_index = 0
```

```python
def reflex_agent(state):
 if state =="Dirty":
    return "clean"

 else:
       return "move"
```

```python
def draw_environment(env, agent_pos, step):
    fig, ax = plt.subplots()
    ax.set_xlim(0, 2)
    ax.set_ylim(0, 2)
    ax.set_xticks([])
    ax.set_yticks([])

    ax.set_title(f"Steps {step} - Agent in {rooms[agent_pos]}")

    for room, pos in room_positions.items():
        x, y = pos
        color = 'red' if env[room] == "Dirty" else 'green'
        rect = patches.Rectangle((x, y), 1, 1, facecolor=color, edgecolor='black')
        ax.add_patch(rect)
        ax.text(x + 0.5, y + 0.5, room, ha='center', va='center', color='white')



      # Draw agent
    agent_x, agent_y = room_positions[rooms[agent_pos]]
    agent_circle = patches.Circle((agent_x + 0.5, agent_y + 0.5), 0.15, color='blue')
    ax.add_patch(agent_circle)

    plt.pause(1)
    plt.close()

```

```python

# Simulation loop
agent_position = 0

for step in range(8):
    current_room = rooms[agent_position]

    # If dirty â†’ clean it
    if environment[current_room] == "Dirty":
        environment[current_room] = "Clean"
    else:
        # Move to next room
        agent_position = (agent_position + 1) % 4

    draw_environment(environment, agent_position, step)
```
